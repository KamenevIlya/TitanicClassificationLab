### Предобработка данных

```python
data.info()
```

Начнем с того, что в некоторых колонках у нас имеются пропущенные значения

Значение Embarked пропущено только у двух пассажиров, поэтому эти записи можно спокойно удалить

```python
data.dropna(axis=0, subset=["Embarked"], inplace=True)
```

Значение Age пропущено у 177 человек, можно попытаться его восстановить. Для простоты заменим его медианным значением

```python
data.fillna({"Age": data["Age"].median()}, axis=0, inplace=True)
```

Большое количество пропусков в поле Cabin. 

```python
data.groupby("Pclass").agg({"Cabin": "count"}).rename(columns={"Cabin": "CabinIsNotNaN"})
```

Можно предположить, что каюты были, в большинстве своем, только у пассажиров первого класса. Роэтому преобразуем данный признак в "Наличие каюты"

```python
data["CabinAvailability"] = 1
data["CabinAvailability"][data["Cabin"].isna()] = 0
data.drop(columns="Cabin", inplace=True)
```

```python
data["CabinAvailability"] = data["CabinAvailability"].astype("category")
```

Признаки Ticket и Name являются текстовыми, и навряд ли полезны в решении нашей задачи, поэтому их мы удалим

```python
data.drop(columns=["Ticket", "Name"], inplace=True)
```

```python
data.info()
```

```python
data.describe(include="all")
```

Т.к. мы будем использовать алгоритмы, использующие в своей работе метрики расстояния, то необходимо нормализовать числовые признаки

```python
data_numerical = data[numeric_col]
data_numerical = (data_numerical - data_numerical.mean(axis = 0))/data_numerical.std(axis = 0)
data[numeric_col] = data_numerical
```

Категорильные признаки с друмя уникальными значениями биноризуем

```python
from sklearn.preprocessing import LabelEncoder

binary_categorial_col = ["Survived", "Sex", "CabinAvailability"]

for col in binary_categorial_col:
    le = LabelEncoder()
    le.fit(data[col])
    data[col] = le.transform(data[col]).astype("float64")
```

```python
data.describe()
```

Категориальные признаки с более чем 2-мя значениями векторизуем:

```python
non_binary_categorial_col = ["Pclass", "Embarked"]
data = data.merge(right=pd.get_dummies(data[non_binary_categorial_col]), left_index=True, right_index=True)
data = data.drop(columns=["Pclass", "Embarked"])
```

```python
data.head()
```

```python
%pip install imbalanced-learn
```

Осталось побороть несбалансированность классов, но для этого нужно разбить выборку X и y, а затем применить adasyn

```python
from imblearn.over_sampling import ADASYN
from collections import Counter

X = data.drop(columns="Survived")
y = data.Survived

print(f'Original dataset shape {Counter(y)}')

ada = ADASYN(random_state=7)

X, y = ada.fit_resample(X, y)
print(f'Resampled dataset shape {Counter(y)}')
```

### Разбиение на обучающую и тестовую выборки

Разобъем данные в отношении 2 к 1

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 7)
```

## Обучение моделей

Т.к. проблема несбалансированности классов решена, то мы имеем полное право в качестве метрики качества взять accuracy

```python
from sklearn.metrics import confusion_matrix

def get_score_info(clf, X_train, X_test, y_train, y_test):

    y_train_prediction = clf.predict(X_train)
    y_test_prediction = clf.predict(X_test)

    print(f'Accuracy на обучающей выборке: {clf.score(X_train, y_train)}')
    print(f'Accuracy на тестовой выборке: {clf.score(X_test, y_test)}')

    print("Матрица рассогласования на тестовой выборке: ")
    print(confusion_matrix(y_test, y_test_prediction))
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

parametrs = {"n_neighbors":range(1, 5)}
neighbor_classifier = KNeighborsClassifier()

grid_search = GridSearchCV(estimator=neighbor_classifier, param_grid=parametrs, scoring="accuracy", cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
best_clf = grid_search.best_estimator_
best_clf.fit(X_train, y_train)

get_score_info(best_clf, X_train, X_test, y_train, y_test)
```

### Decision tree

```python
from sklearn.tree import DecisionTreeClassifier

parametrs = {"max_depth":range(3, 10)}
tree_classifier = DecisionTreeClassifier()

grid_search = GridSearchCV(estimator=tree_classifier, param_grid=parametrs, scoring="accuracy", cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
best_clf = grid_search.best_estimator_
best_clf.fit(X_train, y_train)

get_score_info(best_clf, X_train, X_test, y_train, y_test)
```

### Logistic regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

logistic_clf = LogisticRegression()
logistic_clf.fit(X_train, y_train)

get_score_info(logistic_clf, X_train, X_test, y_train, y_test)
```

### Random forest

```python
from sklearn.ensemble import RandomForestClassifier

parametrs = {"n_estimators":range(10, 200, 50), "max_depth":range(3, 10)}
forest_classifier = RandomForestClassifier()

grid_search = GridSearchCV(estimator=forest_classifier, param_grid=parametrs, scoring="accuracy", cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
best_clf = grid_search.best_estimator_
best_clf.fit(X_train, y_train)

get_score_info(best_clf, X_train, X_test, y_train, y_test)
```

### MLP classifier

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

alpha_arr = np.logspace(-3, 2, 20)

train_acc = []
test_acc = []

for alpha in alpha_arr:
    mlp_model = MLPClassifier(alpha=alpha,
                              hidden_layer_sizes=(100, ),
                              solver='adam',
                              activation='relu',
                              max_iter=1500,
                              random_state=7
                              )
    mlp_model.fit(X_train, y_train)

    y_train_pred = mlp_model.predict(X_train)
    y_test_pred = mlp_model.predict(X_test)

    train_acc.append(accuracy_score(y_train, y_train_pred))
    test_acc.append(accuracy_score(y_test, y_test_pred))
```

```python
plt.semilogx(alpha_arr, train_acc, 'b-o', label="train")
plt.semilogx(alpha_arr, test_acc, 'r-o', label="test")
plt.xlim([np.min(alpha_arr), np.max(alpha_arr)])
plt.title('accuracy vs. alpha')
plt.xlabel('alpha')
plt.ylabel('accuracy')
plt.legend()
```

```python
parametrs = {"alpha":np.logspace(-3, 2, 20)}
mlp_classifier = MLPClassifier(hidden_layer_sizes=(100, ), solver='adam', activation='relu', max_iter=1500, random_state=7)

grid_search = GridSearchCV(estimator=mlp_classifier, param_grid=parametrs, scoring="accuracy", cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
best_clf = grid_search.best_estimator_
best_clf.fit(X_train, y_train)

get_score_info(best_clf, X_train, X_test, y_train, y_test)
```
